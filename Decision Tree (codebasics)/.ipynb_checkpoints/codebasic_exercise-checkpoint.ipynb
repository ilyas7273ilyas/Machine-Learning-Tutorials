# codebasics solution by splitting training and testing data data into 70:30

```python
import pandas as pd
df = pd.read_csv('C:/Users/ilyas/Desktop/Jupitor/Machine Learning/codebasics_materials/py-master/ML/9_decision_tree/Exercise/titanic.csv')
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
df.head()
```

```python
inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
```

```python
inputs.head()
```

```python
# Missing Value imputation for Age column
inputs.Age = inputs.Age.fillna(inputs.Age.mean())  #Filling them with avg value
inputs
```

```python
# Training the data after splitting into 70:30
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs,target,test_size = 0.3)
```

```python
len(x_train)
```

```python
len(y_test)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
# Now predict the survival of a person who is male(1),Pclass is 3, 
 # Age is 35.0, Fare is 8.0500 (4th person, answer should be 0)
model.predict([[3,35.000000,8.0500,1]])
```
